```python
import pandas as pd
```

```python
df=pd.read_csv("train.csv.csv")
df.head()
print(df.shape)
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df["Age"].fillna(df["Age"].median(),inplace=True)
```

```python
df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
```

```python
df.drop(columns=["Cabin"],inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
df["Sex"]=df["Sex"].str.lower().str.strip()
df["Embarked"]=df["Embarked"].str.upper().str.strip()
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
print(df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Survived",data=df)
plt.title("Survival count")
plt.xlabel("Survived(0=No,1=Yes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sex",hue="Survived",data=df)
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Passenger Class vs Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df["Fare"], bins=30, kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Observations:")
print("1. More passengers died than survived.")
print("2. Females had a higher survival rate than males.")
print("3. First-class passengers survived more than lower classes.")
print("4. Most passengers were aged between 20 and 40.")
print("5. Fare and passenger class influenced survival chances.")
```
